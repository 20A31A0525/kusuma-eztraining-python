```python
q1="""Who is developed the python programming language
     a) Wick van Rossum
     b)Rasmus Lerdorf
     c)Guido van Rossum
     d) Niene Stom"""
  
q2=""" Which of the following is the correct extension
       of the python file?
       a).python
       b).pl
       c).py
       d).p"""
q3="""What will be the value of the following python expression?
                   4+3%5
      a)7
      b)2
      c)4
      d)1"""
q4="""which of the following is used to give single-line comments in python?
      a)//
      b)#
      c)!
      d)/*"""
q5="""what will be the output of the following pyhton code snippet if x=1?
                    x<<2
      a) 4
      b) 2
      c) 1
      d) 8"""
question={q1:'c',q2:'c',q3:'a',q4:'b',q5:'a'}
name=input("Hi What is your good name?")
print("Hello ",name,"Welcome to quiz!")
score=0
for i in question:
  print()
  print(i)
  c=input("Do you want to skip the question (yes/no)")
  if(c=='yes'):
    continue
  d=input("enter your answer:")
  if d==question[i]:
    score+=1
    print("Wow you got one point")
    print("your total score is:",score)
  else:
    print("Wrong answer u lost one point")
    score-=1
    print("your total score is:",score)
  s=input("do you want to quit the quiz")
  if s=='yes':
    break
print("Your total score is ",score)
```
